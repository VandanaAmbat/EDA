5.Find if there is any correlation between age and salary , represent it visually.

```python
df2
```

```python
df4=df2.drop(columns=['Name','Team','Position','Height','Number'],axis=1)
df4
```

```python
corr=df4.corr()
print(corr)
```

<font color=Blue>Data Story: There exist a positive correlation between Age and Salary where the values lies between 0 and 1.But it is evident from the results that the correlation is not really very high.</font>

```python
sns.heatmap(corr,annot=True,cmap='flare')
plt.title("Correlation between Age and Salary")
plt.show()
```

<font color=Blue>Data story: The graphical representation of Correlation between age and salary is shown in the above graph. Seaborn is used here for developing this graph.</font>
